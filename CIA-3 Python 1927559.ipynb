# 3. Data Cleaning

```python
# Dropping Unnecessary columns with NA values like seen above 
temp=['ADDRESSLINE1','ADDRESSLINE2','POSTALCODE', 'TERRITORY', 'PHONE', 'CITY' , 'STATE']
data.drop(temp,axis=1,inplace=True)
```

```python
data.isnull().sum()
# Now we see we have no null or NA values
```

```python
#Checking outliers using boxplot using matplotlib library
plt.rcParams['figure.figsize'] = [18, 16]
data.plot(kind="box",subplots=True,layout=(4,4),sharex=False,sharey=False)
plt.show()
```

```python
#scatter plot using Pandas Data Visualization with MSRP column as color map
data.plot.scatter(x='QUANTITYORDERED',y='SALES',c='MSRP',cmap='coolwarm')
```

```python
data.plot.scatter(x='QUANTITYORDERED',y='SALES')
```

# 4. Data Visualization

```python
#density plots to see the different variables and their sales
plt.rcParams['figure.figsize'] = [18, 16]
data.plot(kind="density",subplots=True,layout=(4,4),sharex=False,sharey=False)
plt.show()
```

```python
#Creating a correlation heatmap using seaborn library to check correlation between different variables
sns.heatmap(data.corr(), annot = True, vmin=-1, vmax=1, center= 0)
```

# 5. Deriving useful Insights from the above chart:- 

1. Most of data for year 2003,2004(YEAR_ID). 4th Quater have more sale... 4>1>2>3...
2. According to data, most of sales are with in particular price range. But we have some outliers within 'SALES' and 'QUANTITYORDERED'.
3. We have variables with high variance like 'PRICEEACH','ORDERLINENUMBER' and 'MSRP'.

```python
#Country wise sales:
(data.groupby('COUNTRY').sum()['SALES']).sort_values()
```
